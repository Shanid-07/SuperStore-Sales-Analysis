```python
import matplotlib.pyplot as plt
y=df.loc[0:80,'Profit']
plt.plot(y)
plt.show()
```

```python
gen=df['Category'].value_counts()
print(type(gen))
print(gen)
x=gen.values
lb=gen.index
plt.figure(figsize=(8,6),dpi=100)
plt.pie(x,labels=lb)
plt.legend(lb)
plt.title("Category")
plt.show()
```

```python
popular_pizzas = df.groupby('Product Name')['Quantity'].sum().head(10)
popular_pizzas.plot(kind='bar')
plt.title("Top 10 Best-Selling Products")
plt.ylabel("Total Quantity Sold")
```

```python
category_revenue = df.groupby('Category')['Profit'].sum().sort_values()
category_revenue.plot(kind='barh', color='skyblue')
plt.title("Revenue by  Category")
plt.xlabel("Total Revenue")
```
